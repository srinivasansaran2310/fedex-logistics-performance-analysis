## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
fedex_df.columns
````

````python
# Identifying duplicate records in the dataset

duplicate_values = fedex_df.duplicated().sum()
print("Duplicate Values :", duplicate_values)
````

````python
#Convert Date Columns
date_columns = ["PQ First Sent to Client Date", "PO Sent to Vendor Date",
             "Scheduled Delivery Date", "Delivered to Client Date"]

for col in date_columns:
    if col in fedex_df.columns:
        fedex_df[col] = pd.to_datetime(fedex_df[col], errors='coerce')
````

````python
pd.set_option('display.max_columns', None)
````

````python
#Convert Numeric Columns
numeric_columns = ["line_item_quantity", "line_item_value",
            "pack_price", "unit_price", "weight_kilograms", "freight_cost_usd"]

for col in numeric_columns:
    if col in fedex_df.columns:
        fedex_df[col] = pd.to_numeric(fedex_df[col], errors="coerce")
````

````python
# Geting the most common value
shipment_mode = fedex_df['Shipment Mode'].mode()[0]

# Filling with the mode
fedex_df['Shipment Mode'].fillna(shipment_mode, inplace=True)

# Filling missing Line Item insurance values with 0
fedex_df['Line Item Insurance (USD)'] = fedex_df['Line Item Insurance (USD)'].fillna(0)

# Filling missing values in 'Dosage' column
# Replacing missing dosage values with 'Not Available'

fedex_df['Dosage'].fillna('Not Available', inplace=True)

# Filling missing values in
# 'Line Item Insurance (USD)' column with median value

fedex_df['Line Item Insurance (USD)'].fillna(
    fedex_df['Line Item Insurance (USD)'].median(),
    inplace=True
)
````

````python
# Delivery Lead Time
# Difference between scheduled and delivered dates

fedex_df['Delivery Lead Time'] = (
    fedex_df['Delivered to Client Date'] -
    fedex_df['Scheduled Delivery Date']
).dt.days

# Delay Status
# Identifying delayed shipments

fedex_df['Delayed Shipment'] = np.where(
   fedex_df['Delivery Lead Time'] > 0,
    'Delayed',
    'On Time'
)
````

### What all manipulations have you done and insights you found?

During data wrangling, duplicate values were checked and no duplicate records were found in the dataset. Date columns were converted into datetime format, and important numerical columns were converted into numeric datatype for proper analysis.

Missing values in the 'Shipment Mode', 'Dosage', and 'Line Item Insurance (USD)' columns were handled using mode, median, and suitable replacement values. New KPI columns such as 'Delivery Lead Time' and 'Delayed Shipment' were also created to analyze delivery performance.

The preprocessing helped improve data quality and enabled better analysis of shipment delays, transportation costs, and logistics efficiency.

#### Chart - 2

````python

# Chart - 2 : Delayed vs On-Time Shipments
# ==========================================

fig, ax = plt.subplots(figsize=(7,7))

# Calculate shipment status counts
delay_counts = fedex_df['Delayed Shipment'].value_counts()

# Total shipments
total_shipments = len(fedex_df)

# Calculate percentages
percentages = (delay_counts / total_shipments) * 100

# Create explode effect for largest segment
explode = [0.1 if i == 0 else 0 for i in range(len(delay_counts))]

# Create pie chart
wedges, texts, autotexts = ax.pie(
    delay_counts.values,
    labels=delay_counts.index,
    explode=explode,
    autopct='%.0f%%',
    startangle=90,
    colors=sns.color_palette('Set2'),
    textprops={'fontsize':11, 'fontweight':'bold'},
    wedgeprops={'edgecolor':'black', 'linewidth':1, 'alpha':0.9}
)

# Styling percentage labels
for autotext in autotexts:
    autotext.set_color('darkblue')
    autotext.set_fontweight('bold')
    autotext.set_fontsize(10)

# Chart title
plt.title(
    'Distribution of Delayed and On-Time Shipments\n',
    fontsize=16,
    fontweight='bold',
    color='#2E4053',
    pad=20
)

# Legend labels with count and percentage
legend_labels = [
    f'{status}: {count:,} shipments ({percent:.1f}%)'
    for status, count, percent in zip(
        delay_counts.index,
        delay_counts.values,
        percentages
    )
]

# Adding legend
plt.legend(
    wedges,
    legend_labels,
    title='Shipment Status',
    loc='lower right',
    bbox_to_anchor=(1.3, 0),
    fontsize=9,
    title_fontsize=10
)

# Adjust layout
plt.tight_layout()

# Display chart
plt.show()
````

##### 1. Why did you pick the specific chart?

A pie chart was used because it clearly shows the proportion of delayed and on-time shipments, making it easy to compare shipment performance distribution.

##### 2. What is/are the insight(s) found from the chart?

The chart shows that most shipments were delivered on time, while a smaller percentage experienced delays. This indicates overall good delivery performance with some operational delays.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Positive Impact:

A high percentage of on-time shipments reflects efficient logistics operations and improves customer satisfaction and service reliability.

Negative Impact:

The delayed shipments may lead to customer dissatisfaction, increased transportation costs, and reduced operational efficiency if delays continue frequently.
